+ Broadcasting (1-D array)

```python
# array vs scalar
array_1 = np.array([1, 2, 3, 4])
sc = 2
print(f'array_1 + sc = {array_1 + sc}')
print(f'array_1 - sc = {array_1 - sc}')
print(f'array_1 * sc = {array_1 * sc}')
print(f'array_1 / sc = {array_1 / sc}')
print(f'array_1 % sc = {np.mod(array_1, sc)}') # floor division
```

```python
# array vs array
array_1 = np.array([1, 2, 3, 4])
array_2 = np.array([5, 6, 7, 8])
print(f'array_1 + array_2 = {array_1 + array_2}')
print(f'array_1 - array_2 = {array_1 - array_2}')
print(f'array_1 * array_2 = {array_1 * array_2}')
print(f'array_1 / array_2 = {array_1 / array_2}')
print(f'array_1 % array_2 = {np.mod(array_1, array_2)}') # floor division
```

+ statistical function

```python
# mean, median
array_2 = np.array([5, 6, 7, 8, 9, 8, 6, 7, 10])
print(f'mean = {np.mean(array_2)}')
print(f'median = {np.median(array_2)}')
```

+ Sort, where, extract

```python
# ascending order --> np.sort()
array_2 = np.array([5, 6, 7, 8, 9, 8, 6, 7, 10])
print(f'array_2 = {array_2}')
print(f'np.sort(array_2) = {np.sort(array_2)}') # ascending
print(f'np.sort(array_2) = {np.sort(array_2)[::-1]}') # descending
```

```python
array_2 = np.array([5, 6, 7, 8, 9, 8, 6, 7, 10])
indexes = np.where(array_2 > 8)
print(f'indexes = {indexes}')
print(f'array_2[indexes] = {array_2[indexes]}')
```

```python
# extract(điều kiện, mảng)
array_2 = np.array([5, 6, 7, 8, 9, 8, 6, 7, 10])
print(f'np.extract(array_2 > 8, array_2) = {np.extract(array_2 > 8, array_2)}')
```

+ statistical function

```python
a = np.array([[1,2,3], [4,5,6]])
print(a)
```

```python
# min; max; mean; median; std (standard deviation)
print(f'max = {np.max(a)}')
print(f'min = {np.min(a)}')
print(f'mean = {np.mean(a)}')
print(f'median = {np.median(a)}')
print(f'std = {np.std(a)}')

print(f'max (axis = 0) = {np.max(a, 0)}')
print(f'max (axis = 1) = {np.max(a, 1)}')
print(f'min (axis = 0) = {np.min(a, 0)}')
print(f'min (axis = 1) = {np.min(a, 1)}')
print(f'mean (axis = 0) = {np.mean(a, 0)}')
print(f'mean (axis = 1) = {np.mean(a, 1)}')
print(f'median (axis = 0) = {np.median(a, 0)}')
print(f'median (axis = 1) = {np.median(a, 1)}')
print(f'std (axis = 0) = {np.std(a, 0)}')
print(f'std (axis = 1) = {np.std(a, 1)}')
```
